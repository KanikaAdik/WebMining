```python
def get_job_details(job_list):
    columns = ['job_title',
               'job_loc',
               'job_salary',
               'job_description']
    temp_df = pd.DataFrame(columns=columns)
    
    for job in job_list:
        job.click()
        time.sleep(5)
        
        dd = driver.find_element(By.CLASS_NAME, "jobsearch-ViewJobLayout-jobDisplay") #finding description layout
        title = dd.find_element(By.CLASS_NAME, "jobsearch-JobInfoHeader-title-container").text #title
        location = dd.find_element(By.CLASS_NAME, "jobsearch-CompanyInfoContainer").text #company
        try:
            salary= dd.find_element(By.ID, "salaryInfoAndJobType").text #css-2iqe2o eu4oa1w0
        except:
            salary=""
            pass
        job_description = dd.find_element(By.ID, "jobDescriptionText").text #salary jobsearch-jobDescriptionText
        #jd = dd.find_element(By.CLASS_NAME, "jobsearch-jobDescriptionText").text
        #print(s, "\njjjdd::", jd)
        #print("\nTitle:", title,"\nlocation", location, "\nsalary",salary,"\njob_description", job_description)
        #temp_df.append({'job_title':title,'job_loc':location,'job_salary':salary,'job_description':job_description})
        temp_df.loc[len(temp_df.index)] = [title,location,salary,job_description] 
        #print(temp_df)
    return temp_df
```

```python
def getRecords(page_url, driver):
    # add your code here
    driver.get(page_url)
    records =[]
    # now get the page content dynamically generated
    page = driver.page_source
    
    # parse the content by beautifulsoup 
    soup = BeautifulSoup(page, 'html.parser')
    #time.sleep(3)
    
    nav = driver.find_elements(By.CSS_SELECTOR,'a[href^="/jobs?q=accountant"]' )
    jj = driver.find_elements(By.CSS_SELECTOR,'a[href^="/rc/clk?jk"]')
    #links = driver.find_elements(By.CSS_SELECTOR,'a[href^="/pagead/clk?"]')
    
    columns = ['job_title',
               'job_loc',
               'job_salary',
               'job_description']
    df = pd.DataFrame(columns=columns)
    print("check for while loop ", len(df), "\n df.index",len(df.index))
    while (len(df.index)<=1000):
        records= get_job_details(jj)
        time.sleep(5)
        nextpg= nav[-1]
        nextpg.click()
        jj = driver.find_elements(By.CSS_SELECTOR,'a[href^="/rc/clk?"]' )
        nav = driver.find_elements(By.CSS_SELECTOR,'a[href^="/jobs?q=accountant"]' )
        #print("\nin loop::::",len(jj), len(nav))
        df = pd.concat([df, records])
        print("len of df::::: ", len(df))
    return df
```
